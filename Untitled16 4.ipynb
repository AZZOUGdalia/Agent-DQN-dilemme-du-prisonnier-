```python
"""
VERSION AMÃ‰LIORÃ‰E (V2.0)
- Ajout d'un Target Network (DQN Stable)
- Ajout d'un graphique de Taux de CoopÃ©ration
- HyperparamÃ¨tres affinÃ©s
"""

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque, defaultdict
import matplotlib.pyplot as plt
import pandas as pd
import copy

# ==================== CONFIGURATION ====================
class Config:
    ACTION_COOP = 0
    ACTION_DEFECT = 1

    # DQN
    STATE_SIZE = 3
    HIDDEN_SIZE = 64
    LEARNING_RATE = 0.001  # Plus lent pour Ãªtre plus prÃ©cis
    GAMMA = 0.95
    MEMORY_SIZE = 50000     # Plus de mÃ©moire
    BATCH_SIZE = 64         # Batch plus gros

    EPSILON_START = 1.0
    EPSILON_END = 0.05      # On garde un peu d'exploration (5%)
    EPSILON_DECAY = 0.9995   # DÃ©croissance trÃ¨s lente

    TARGET_UPDATE = 100     # FrÃ©quence de mise Ã  jour du rÃ©seau cible (steps)

    # SARSA
    SARSA_ALPHA = 0.1
    SARSA_GAMMA = 0.95
    SARSA_EPSILON = 0.1

    # Simulation
    NUM_EPISODES = 600
    STEPS_PER_MATCH = 50

config = Config()

PAYOFF = {
    (0, 0): (3, 3), (0, 1): (0, 5),
    (1, 0): (5, 0), (1, 1): (1, 1)
}

# ==================== CLASSES DES BOTS (InchangÃ©es) ====================

class Competitor:
    def __init__(self, name, id_num):
        self.name = name
        self.id = id_num
    def reset(self): pass
    def act(self, opp_last): return random.choice([0, 1])

class RandomBot(Competitor):
    def act(self, opp_last): return random.choice([0, 1])

class AngelBot(Competitor):
    def act(self, opp_last): return 0

class DevilBot(Competitor):
    def act(self, opp_last): return 1

class TitForTatBot(Competitor):
    def act(self, opp_last):
        return 0 if opp_last is None else opp_last

class SarsaBot(Competitor):
    def __init__(self, name, id_num):
        super().__init__(name, id_num)
        self.q_table = defaultdict(lambda: [0.0, 0.0])
        self.last_state = (0, 0)
        self.last_action = 0

    def reset(self):
        self.last_state = (0, 0)
        self.last_action = 0

    def act(self, opp_last):
        current_state = (self.last_action, opp_last if opp_last is not None else 0)
        if random.random() < config.SARSA_EPSILON:
            action = random.choice([0, 1])
        else:
            qs = self.q_table[current_state]
            action = 0 if qs[0] > qs[1] else 1
        self.current_state_buffer = current_state
        self.current_action_buffer = action
        return action

    def learn(self, reward, next_opp_action):
        next_state = (self.current_action_buffer, next_opp_action)
        old_q = self.q_table[self.current_state_buffer][self.current_action_buffer]
        max_next_q = max(self.q_table[next_state])
        new_q = old_q + config.SARSA_ALPHA * (reward + config.SARSA_GAMMA * max_next_q - old_q)
        self.q_table[self.current_state_buffer][self.current_action_buffer] = new_q
        self.last_action = self.current_action_buffer
        self.last_state = next_state

# ==================== AGENT DQN AMÃ‰LIORÃ‰ (Target Network) ====================
class DQNAgent(nn.Module):
    def __init__(self):
        super(DQNAgent, self).__init__()

        # 1. Policy Network (Celui qui apprend vite)
        self.policy_net = nn.Sequential(
            nn.Linear(config.STATE_SIZE, config.HIDDEN_SIZE),
            nn.ReLU(),
            nn.Linear(config.HIDDEN_SIZE, config.HIDDEN_SIZE),
            nn.ReLU(),
            nn.Linear(config.HIDDEN_SIZE, 2)
        )

        # 2. Target Network (La copie stable)
        self.target_net = copy.deepcopy(self.policy_net)
        self.target_net.eval() # Pas de gradients pour lui

        self.optimizer = optim.Adam(self.policy_net.parameters(), lr=config.LEARNING_RATE)
        self.memory = deque(maxlen=config.MEMORY_SIZE)
        self.epsilon = config.EPSILON_START
        self.steps_done = 0

    def select_action(self, state):
        self.steps_done += 1
        if random.random() < self.epsilon:
            return random.choice([0, 1])
        with torch.no_grad():
            # On utilise le Policy Net pour choisir l'action
            return self.policy_net(torch.FloatTensor(state)).argmax().item()

    def train_step(self):
        if len(self.memory) < config.BATCH_SIZE: return

        batch = random.sample(self.memory, config.BATCH_SIZE)
        states, actions, rewards, next_states = zip(*batch)

        states_t = torch.FloatTensor(np.array(states))
        actions_t = torch.LongTensor(actions).unsqueeze(1)
        rewards_t = torch.FloatTensor(rewards)
        next_states_t = torch.FloatTensor(np.array(next_states))

        # Q(s, a) actuel prÃ©dit par Policy Net
        q_curr = self.policy_net(states_t).gather(1, actions_t).squeeze()

        # Q(s', a') max prÃ©dit par TARGET NET (StabilitÃ© !)
        with torch.no_grad():
            q_next = self.target_net(next_states_t).max(1)[0]

        expected_q = rewards_t + config.GAMMA * q_next

        loss = nn.SmoothL1Loss()(q_curr, expected_q) # Huber Loss (mieux que MSE)

        self.optimizer.zero_grad()
        loss.backward()
        # Clip gradients pour Ã©viter les explosions
        torch.nn.utils.clip_grad_norm_(self.policy_net.parameters(), 1.0)
        self.optimizer.step()

        self.epsilon = max(config.EPSILON_END, self.epsilon * config.EPSILON_DECAY)

        # Mise Ã  jour du Target Network de temps en temps
        if self.steps_done % config.TARGET_UPDATE == 0:
            self.target_net.load_state_dict(self.policy_net.state_dict())

    def remember(self, s, a, r, ns): self.memory.append((s, a, r, ns))


# ==================== LOGIQUE DU TOURNOI ====================
def run_improved_tournament():
    print("ðŸš€ DÃ‰MARRAGE V2.0 : Target Network & MÃ©triques AvancÃ©es")

    dqn = DQNAgent()

    sarsa_bot = SarsaBot("SARSA", 4)
    static_bots = [
        AngelBot("Angel", 0),
        DevilBot("Devil", 1),
        TitForTatBot("TfT", 2),
        RandomBot("Random", 3)
    ]
    all_opponents = static_bots + [sarsa_bot]

    # Historique Scores + Taux de CoopÃ©ration
    history_scores = {bot.name: [] for bot in all_opponents}
    history_scores["DQN"] = []

    history_coop = {bot.name: [] for bot in all_opponents} # Taux de coop de l'ADVERSAIRE
    history_coop["DQN"] = [] # Taux de coop du DQN (Moyenne globale)

    for episode in range(config.NUM_EPISODES):

        # --- PHASE 1 : ENTRAÃŽNEMENT BACKGROUND SARSA ---
        for sparring_partner in static_bots:
            sarsa_bot.reset(); sparring_partner.reset()
            s_last, p_last = 0, 0
            for _ in range(20):
                act_s = sarsa_bot.act(p_last)
                act_p = sparring_partner.act(s_last)
                r_s, _ = PAYOFF[(act_s, act_p)]
                sarsa_bot.learn(r_s, act_p)
                s_last, p_last = act_s, act_p

        # --- PHASE 2 : ROUND ROBIN ---
        dqn_ep_score = 0
        dqn_ep_coop_count = 0
        total_steps_ep = 0

        for opponent in all_opponents:
            opponent.reset()

            opp_score = 0
            opp_coop_count = 0
            dqn_last, opp_last = 0, 0

            for step in range(config.STEPS_PER_MATCH):
                state = np.array([dqn_last, opp_last, opponent.id/4.0], dtype=np.float32)

                act_dqn = dqn.select_action(state)
                act_opp = opponent.act(dqn_last)

                r_dqn, r_opp = PAYOFF[(act_dqn, act_opp)]
                next_state = np.array([act_dqn, act_opp, opponent.id/4.0], dtype=np.float32)

                dqn.remember(state, act_dqn, r_dqn, next_state)
                dqn.train_step()

                dqn_last, opp_last = act_dqn, act_opp

                # MÃ©triques
                opp_score += r_opp
                dqn_ep_score += r_dqn
                if act_opp == 0: opp_coop_count += 1
                if act_dqn == 0: dqn_ep_coop_count += 1

                total_steps_ep += 1

            # Sauvegarde par adversaire
            history_scores[opponent.name].append(opp_score / config.STEPS_PER_MATCH)
            history_coop[opponent.name].append(opp_coop_count / config.STEPS_PER_MATCH)

        # Sauvegarde Globale DQN
        history_scores["DQN"].append(dqn_ep_score / total_steps_ep)
        history_coop["DQN"].append(dqn_ep_coop_count / total_steps_ep)

        if (episode+1) % 50 == 0:
            print(f"Ep {episode+1:3d} | DQN Score: {history_scores['DQN'][-1]:.2f} | DQN Coop: {history_coop['DQN'][-1]*100:.0f}%")

    return history_scores, history_coop

# ==================== VISUALISATION (2 Graphiques) ====================
if __name__ == "__main__":
    scores, coops = run_improved_tournament()

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))

    colors = {"Angel": "cyan", "Devil": "black", "TfT": "blue",
              "Random": "gray", "SARSA": "red", "DQN": "green"}

    # --- GRAPHIQUE 1 : SCORES ---
    for name, data in scores.items():
        smooth = pd.Series(data).rolling(window=20).mean()
        lw = 3 if name == "DQN" else 1.5
        ls = '-' if name == "DQN" else '--'
        ax1.plot(smooth, label=name, color=colors.get(name), linewidth=lw, linestyle=ls)

    ax1.set_title("1. RÃ©compenses Moyennes (StabilitÃ© amÃ©liorÃ©e)")
    ax1.set_ylabel("Score")
    ax1.axhline(3.0, color='green', alpha=0.2); ax1.axhline(1.0, color='red', alpha=0.2)
    ax1.legend(loc='center right')
    ax1.grid(True, alpha=0.3)

    # --- GRAPHIQUE 2 : TAUX DE COOPÃ‰RATION ---
    for name, data in coops.items():
        smooth = pd.Series(data).rolling(window=20).mean()
        lw = 3 if name == "DQN" else 1.5
        ls = '-' if name == "DQN" else ':'
        ax2.plot(smooth, label=name, color=colors.get(name), linewidth=lw, linestyle=ls)

    ax2.set_title("2. Taux de CoopÃ©ration (0 = Trahison Totale, 1 = CoopÃ©ration Totale)")
    ax2.set_ylabel("Taux (0-1)")
    ax2.set_xlabel("Ã‰pisodes")
    ax2.legend(loc='center right')
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
```

```python
"""
CODE FINAL OPTIMISÃ‰ : TOURNOI DQN vs 5 ADVERSAIRES
Protocole :
1. SARSA (Vrai Sarsa On-Policy) s'entraÃ®ne en arriÃ¨re-plan contre les bots statiques.
2. DQN affronte tout le monde.
3. CRUCIAL : Lors du match DQN vs SARSA, SARSA est figÃ© (training=False).
"""

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque, defaultdict
import matplotlib.pyplot as plt
import pandas as pd
import copy

# ==================== 1. CONFIGURATION "CHAMPION" ====================
class Config:
    # --- HYPER-PARAMÃˆTRES (OptimisÃ©s pour la stabilitÃ©) ---
    LR_START = 0.0001       # DÃ©part lent pour ne pas casser la stratÃ©gie
    GAMMA = 0.99            # Vision trÃ¨s long terme (patience)
    MEMORY_SIZE = 50000     # MÃ©moire Ã©norme pour retenir les 5 profils
    BATCH_SIZE = 64         # Batch plus large

    # --- EVOLUTION ---
    EPSILON_START = 1.0
    EPSILON_END = 0.01
    EPSILON_DECAY = 0.999   # Exploration longue

    # SCHEDULER : Le LR sera divisÃ© par 2 tous les 30000 pas
    LR_DECAY_STEP = 30000
    LR_DECAY_GAMMA = 0.5
    TARGET_UPDATE = 200     # FrÃ©quence de mise Ã  jour du Target Net

    # --- SIMULATION ---
    NUM_EPISODES = 1500     # DurÃ©e suffisante pour la convergence
    STEPS_PER_MATCH = 50
    STATE_SIZE = 3          # [MonAction, SonAction, SonID]
    HIDDEN_SIZE = 128       # Cerveau DQN

    # --- SARSA PARAMÃˆTRES ---
    SARSA_ALPHA = 0.1
    SARSA_GAMMA = 0.95
    SARSA_EPSILON = 0.1

config = Config()

PAYOFF = { (0, 0): (3, 3), (0, 1): (0, 5), (1, 0): (5, 0), (1, 1): (1, 1) }

# ==================== 2. CLASSES DES OPPOSANTS ====================

class Competitor:
    def __init__(self, name, id_num): self.name, self.id = name, id_num
    def reset(self): pass
    def act(self, opp_last, training=False): return random.choice([0, 1]) # training arg par compatibilitÃ©

class RandomBot(Competitor):
    def act(self, opp_last, training=False): return random.choice([0, 1])

class AngelBot(Competitor):
    def act(self, opp_last, training=False): return 0

class DevilBot(Competitor):
    def act(self, opp_last, training=False): return 1

class TitForTatBot(Competitor):
    def act(self, opp_last, training=False): return 0 if opp_last is None else opp_last

class SarsaBot(Competitor):
    def __init__(self, name, id_num):
        super().__init__(name, id_num)
        self.q_table = defaultdict(lambda: [0.0, 0.0])
        self.prev_state = None
        self.prev_action = None
        self.prev_reward = None
        self.curr_state = None
        self.curr_action = None

    def reset(self):
        self.prev_state = None
        self.prev_action = None
        self.prev_reward = None
        self.curr_state = (0, 0)
        self.curr_action = 0

    def act(self, opp_last, training=True):
        """
        training=True  : Applique l'Ã©quation de mise Ã  jour SARSA.
        training=False : Choisit juste l'action (Mode Test/FigÃ©).
        """
        # 1. Ã‰tat Actuel S_t
        self.curr_state = (self.curr_action, opp_last if opp_last is not None else 0)

        # 2. Choix Action A_t
        if random.random() < config.SARSA_EPSILON:
            self.curr_action = random.choice([0, 1])
        else:
            qs = self.q_table[self.curr_state]
            self.curr_action = 0 if qs[0] > qs[1] else 1

        # 3. MISE Ã€ JOUR SARSA (Seulement si training=True)
        # Q(S,A) <- Q(S,A) + alpha * [ R + gamma * Q(S',A') - Q(S,A) ]
        if training and self.prev_state is not None and self.prev_reward is not None:
            old_q = self.q_table[self.prev_state][self.prev_action]
            next_q = self.q_table[self.curr_state][self.curr_action] # On-Policy : Q(S', A')

            new_q = old_q + config.SARSA_ALPHA * (self.prev_reward + config.SARSA_GAMMA * next_q - old_q)
            self.q_table[self.prev_state][self.prev_action] = new_q

        # 4. Glissement temporel
        self.prev_state = self.curr_state
        self.prev_action = self.curr_action

        return self.curr_action

    def learn(self, reward):
        # On stocke juste la rÃ©compense pour le prochain appel de act()
        self.prev_reward = reward

# ==================== 3. AGENT DQN (AVANCÃ‰) ====================
class DQNAgent(nn.Module):
    def __init__(self):
        super(DQNAgent, self).__init__()

        self.policy_net = nn.Sequential(
            nn.Linear(config.STATE_SIZE, config.HIDDEN_SIZE), nn.ReLU(),
            nn.Linear(config.HIDDEN_SIZE, config.HIDDEN_SIZE), nn.ReLU(),
            nn.Linear(config.HIDDEN_SIZE, 2)
        )
        self.target_net = copy.deepcopy(self.policy_net); self.target_net.eval()

        self.optimizer = optim.Adam(self.policy_net.parameters(), lr=config.LR_START)
        # Le Scheduler rÃ©duit le LR progressivement
        self.scheduler = optim.lr_scheduler.StepLR(self.optimizer, step_size=config.LR_DECAY_STEP, gamma=config.LR_DECAY_GAMMA)

        self.memory = deque(maxlen=config.MEMORY_SIZE)
        self.epsilon = config.EPSILON_START
        self.steps_done = 0

    def select_action(self, state):
        self.steps_done += 1
        if random.random() < self.epsilon: return random.choice([0, 1])
        with torch.no_grad(): return self.policy_net(torch.FloatTensor(state)).argmax().item()

    def train_step(self):
        if len(self.memory) < config.BATCH_SIZE: return

        batch = random.sample(self.memory, config.BATCH_SIZE)
        s, a, r, ns = zip(*batch)

        q_curr = self.policy_net(torch.FloatTensor(np.array(s))).gather(1, torch.LongTensor(a).unsqueeze(1)).squeeze()
        with torch.no_grad():
            q_next = self.target_net(torch.FloatTensor(np.array(ns))).max(1)[0]

        # SmoothL1Loss
        loss = nn.SmoothL1Loss()(q_curr, torch.FloatTensor(r) + config.GAMMA * q_next)

        self.optimizer.zero_grad(); loss.backward()
        torch.nn.utils.clip_grad_norm_(self.policy_net.parameters(), 1.0)
        self.optimizer.step()
        self.scheduler.step()

        if self.steps_done % config.TARGET_UPDATE == 0:
            self.target_net.load_state_dict(self.policy_net.state_dict())

        self.epsilon = max(config.EPSILON_END, self.epsilon * config.EPSILON_DECAY)

    def remember(self, s, a, r, ns): self.memory.append((s, a, r, ns))

# ==================== 4. SIMULATION ====================
def run_simulation():
    print("ðŸš€ DÃ‰MARRAGE : Vrai SARSA (On-Policy) + DQN Robuste")

    dqn = DQNAgent()
    sarsa_bot = SarsaBot("SARSA", 4)
    bots = [AngelBot("Angel",0), DevilBot("Devil",1), TitForTatBot("TfT",2), RandomBot("Random",3)]
    opponents = bots + [sarsa_bot]

    hist_score = {b.name: [] for b in opponents}; hist_score["DQN"] = []
    hist_coop = {b.name: [] for b in opponents}

    for episode in range(config.NUM_EPISODES):

        # --- PHASE 1 : ENTRAÃŽNEMENT BACKGROUND SARSA ---
        for b in bots:
            sarsa_bot.reset(); b.reset(); sl, pl = 0,0
            for _ in range(20):
                # training=True -> SARSA apprend
                as_ = sarsa_bot.act(pl, training=True)
                ap = b.act(sl)
                rs, _ = PAYOFF[(as_, ap)]
                sarsa_bot.learn(rs) # Passe la rÃ©compense
                sl, pl = as_, ap

        # --- PHASE 2 : TOURNOI DQN ---
        dqn_ep_score = 0

        for opp in opponents:
            opp.reset(); dl, ol = 0,0
            match_score, match_coop = 0, 0

            # DÃ©tection si c'est SARSA pour figer l'apprentissage
            is_sarsa = (opp.name == "SARSA")

            for _ in range(config.STEPS_PER_MATCH):
                s = np.array([dl, ol, opp.id/4.0], dtype=np.float32)

                ad = dqn.select_action(s)

                # ACTION OPPOSANT
                if is_sarsa:
                    ao = opp.act(dl, training=False) # INTERDIT D'APPRENDRE !
                else:
                    ao = opp.act(dl)

                rd, ro = PAYOFF[(ad, ao)]
                ns = np.array([ad, ao, opp.id/4.0], dtype=np.float32)

                # DQN apprend TOUJOURS
                dqn.remember(s, ad, rd, ns)
                dqn.train_step()

                # On n'appelle PAS sarsa.learn() ici

                dl, ol = ad, ao
                match_score += rd
                if ao == 0: match_coop += 1

            hist_score[opp.name].append(match_score / config.STEPS_PER_MATCH)
            hist_coop[opp.name].append(match_coop / config.STEPS_PER_MATCH)
            dqn_ep_score += match_score

        hist_score["DQN"].append((dqn_ep_score/5) / config.STEPS_PER_MATCH)

        if (episode+1) % 100 == 0:
            print(f"Ep {episode+1:4d} | DQN Score: {hist_score['DQN'][-1]:.2f} | Epsilon: {dqn.epsilon:.3f}")

    return hist_score, hist_coop

# ==================== 5. VISUALISATION ====================
if __name__ == "__main__":
    scores, coops = run_simulation()

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12))
    colors = {"Angel":"cyan", "Devil":"black", "TfT":"blue", "Random":"gray", "SARSA":"red", "DQN":"green"}

    # Plot 1 : Scores
    for name, data in scores.items():
        smooth = pd.Series(data).rolling(window=40).mean()
        lw = 3 if name == "DQN" else 1.5; ls = '-' if name == "DQN" else '--'
        ax1.plot(smooth, label=name, c=colors.get(name), lw=lw, ls=ls)
    ax1.set_title("Scores Moyens (DQN apprend Ã  exploiter)"); ax1.legend(); ax1.grid(True, alpha=0.3)
    ax1.axhline(3.0, c='green', alpha=0.2); ax1.axhline(1.0, c='red', alpha=0.2)


    # Plot 2 : CoopÃ©ration
    for name, data in coops.items():
        smooth = pd.Series(data).rolling(window=40).mean()
        ax2.plot(smooth, label=f"Coop de {name}", c=colors.get(name), lw=1.5)
    ax2.set_title("Comportement des Adversaires (Est-ce qu'ils coopÃ¨rent ?)")
    ax2.set_ylabel("Taux Coop (0=Trahison, 1=Amour)"); ax2.legend(); ax2.grid(True, alpha=0.3)


    plt.tight_layout()
    plt.show()
```

```python
"""
GRID SEARCH FINAL (V3.1) : Optimisation avec ParamÃ¨tres Ã‰volutifs
Objectif : Trouver le profil d'Ã©volution parfait (DÃ©part -> Fin).
Protocole : AsymÃ©trique Strict (Sarsa figÃ© contre DQN).
"""

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque, defaultdict
import itertools
import time
import copy

# ==================== 1. LES "BIG 5" Ã€ TESTER ====================
PARAM_GRID = {
    # 1. Le Cerveau (DÃ©part)
    "LR_START":         [0.001, 0.0001],     # Rapide vs Chirurgical

    # 2. L'Ã‰volution (Vitesse de refroidissement du LR)
    # 0.5 = On divise le LR par 2 tous les X pas (Chute brutale)
    # 0.9 = On garde 90% du LR (Chute douce)
    "LR_DECAY_GAMMA":   [0.5, 0.8],

    # 3. La CuriositÃ© (Vitesse de fin d'exploration)
    # 0.990 = Rapide (Fin de l'exploration vers l'Ã©pisode 300)
    # 0.995 = Lent (Fin de l'exploration vers l'Ã©pisode 800)
    "EPSILON_DECAY":    [0.990, 0.995],

    # 4. La Patience
    "GAMMA":            [0.95, 0.99],

    # 5. La MÃ©moire
    "MEMORY_SIZE":      [10000, 50000]
}

# ParamÃ¨tres Fixes (On ne les touche pas pour ce test)
CONSTANTS = {
    "STATE_SIZE": 3,
    "HIDDEN_SIZE": 64,
    "BATCH_SIZE": 64,
    "EPSILON_START": 1.0,
    "EPSILON_END": 0.01,
    "TARGET_UPDATE": 100,
    "LR_DECAY_STEP": 10000, # On rÃ©duit le LR tous les 10k pas (~ tous les 40 Ã©pisodes)

    # Simulation raccourcie pour le tuning
    "NUM_EPISODES_TEST": 350,
    "STEPS_PER_MATCH": 50,

    # SARSA Fixed
    "SARSA_ALPHA": 0.1,
    "SARSA_GAMMA": 0.95,
    "SARSA_EPSILON": 0.1
}

PAYOFF = { (0, 0): (3, 3), (0, 1): (0, 5), (1, 0): (5, 0), (1, 1): (1, 1) }

# ==================== 2. CLASSES ADVERSAIRES (SARSA CORRIGÃ‰) ====================

class Competitor:
    def __init__(self, name, id_num): self.name, self.id = name, id_num
    def reset(self): pass
    def act(self, o, training=False): return random.choice([0, 1])

class RandomBot(Competitor):
    def act(self, o, training=False): return random.choice([0, 1])
class AngelBot(Competitor):
    def act(self, o, training=False): return 0
class DevilBot(Competitor):
    def act(self, o, training=False): return 1
class TitForTatBot(Competitor):
    def act(self, o, training=False): return 0 if o is None else o

class SarsaBot(Competitor):
    def __init__(self, name, id_num):
        super().__init__(name, id_num)
        self.q_table = defaultdict(lambda: [0.0, 0.0])
        self.prev_state = None; self.prev_action = None; self.prev_reward = None
        self.curr_state = None; self.curr_action = None

    def reset(self):
        self.prev_state = None; self.prev_action = None; self.prev_reward = None
        self.curr_state = (0, 0); self.curr_action = 0

    def act(self, opp_last, training=True):
        self.curr_state = (self.curr_action, opp_last if opp_last is not None else 0)

        if random.random() < CONSTANTS["SARSA_EPSILON"]: self.curr_action = random.choice([0, 1])
        else:
            qs = self.q_table[self.curr_state]
            self.curr_action = 0 if qs[0] > qs[1] else 1

        # Vrai SARSA (On-Policy) + Protection Training
        if training and self.prev_state is not None and self.prev_reward is not None:
            old_q = self.q_table[self.prev_state][self.prev_action]
            next_q = self.q_table[self.curr_state][self.curr_action]
            new_q = old_q + CONSTANTS["SARSA_ALPHA"] * (self.prev_reward + CONSTANTS["SARSA_GAMMA"] * next_q - old_q)
            self.q_table[self.prev_state][self.prev_action] = new_q

        self.prev_state = self.curr_state
        self.prev_action = self.curr_action
        return self.curr_action

    def learn(self, reward): self.prev_reward = reward

# ==================== 3. AGENT DQN (AVEC LES PARAMÃˆTRES Ã‰VOLUTIFS) ====================
class DQNAgent(nn.Module):
    def __init__(self, params):
        super(DQNAgent, self).__init__()
        self.params = params

        self.p_net = nn.Sequential(
            nn.Linear(CONSTANTS["STATE_SIZE"], CONSTANTS["HIDDEN_SIZE"]), nn.ReLU(),
            nn.Linear(CONSTANTS["HIDDEN_SIZE"], CONSTANTS["HIDDEN_SIZE"]), nn.ReLU(),
            nn.Linear(CONSTANTS["HIDDEN_SIZE"], 2)
        )
        self.t_net = copy.deepcopy(self.p_net); self.t_net.eval()

        # 1. Utilisation du LR_START
        self.opt = optim.Adam(self.p_net.parameters(), lr=params["LR_START"])

        # 2. Utilisation du LR_DECAY_GAMMA
        self.scheduler = optim.lr_scheduler.StepLR(self.opt, step_size=CONSTANTS["LR_DECAY_STEP"], gamma=params["LR_DECAY_GAMMA"])

        self.mem = deque(maxlen=params["MEMORY_SIZE"])
        self.eps = CONSTANTS["EPSILON_START"]
        self.steps = 0

    def select_action(self, s):
        self.steps += 1
        if random.random() < self.eps: return random.choice([0, 1])
        with torch.no_grad(): return self.p_net(torch.FloatTensor(s)).argmax().item()

    def train_step(self):
        if len(self.mem) < CONSTANTS["BATCH_SIZE"]: return
        batch = random.sample(self.mem, CONSTANTS["BATCH_SIZE"])
        s, a, r, ns = zip(*batch)

        q = self.p_net(torch.FloatTensor(np.array(s))).gather(1, torch.LongTensor(a).unsqueeze(1)).squeeze()
        with torch.no_grad(): qn = self.t_net(torch.FloatTensor(np.array(ns))).max(1)[0]

        loss = nn.SmoothL1Loss()(q, torch.FloatTensor(r) + self.params["GAMMA"] * qn)

        self.opt.zero_grad(); loss.backward(); self.opt.step()
        self.scheduler.step() # Ã‰volution du LR

        if self.steps % CONSTANTS["TARGET_UPDATE"] == 0: self.t_net.load_state_dict(self.p_net.state_dict())

        # 3. Utilisation de EPSILON_DECAY
        self.eps = max(CONSTANTS["EPSILON_END"], self.eps * self.params["EPSILON_DECAY"])

    def remember(self, s, a, r, ns): self.mem.append((s, a, r, ns))

# ==================== 4. Ã‰VALUATION ====================
def evaluate_config(params):
    dqn = DQNAgent(params)
    sarsa = SarsaBot("SARSA", 4)
    bots = [AngelBot("A",0), DevilBot("D",1), TitForTatBot("T",2), RandomBot("R",3)]
    opponents = bots + [sarsa]

    scores = []

    for ep in range(CONSTANTS["NUM_EPISODES_TEST"]):
        # Train SARSA
        for b in bots:
            sarsa.reset(); b.reset(); sl, pl = 0,0
            for _ in range(20):
                as_ = sarsa.act(pl, training=True)
                ap = b.act(sl); rs, _ = PAYOFF[(as_, ap)]
                sarsa.learn(rs); sl, pl = as_, ap

        # Test DQN
        ep_total = 0
        for opp in opponents:
            opp.reset(); dl, ol = 0,0; m_score = 0
            is_sarsa = (opp.name == "SARSA")

            for _ in range(CONSTANTS["STEPS_PER_MATCH"]):
                s = np.array([dl, ol, opp.id/4.0], dtype=np.float32)
                ad = dqn.select_action(s)

                # Sarsa figÃ© ?
                if is_sarsa: ao = opp.act(dl, training=False)
                else: ao = opp.act(dl)

                rd, ro = PAYOFF[(ad, ao)]
                ns = np.array([ad, ao, opp.id/4.0], dtype=np.float32)

                dqn.remember(s, ad, rd, ns)
                dqn.train_step()

                dl, ol = ad, ao
                m_score += rd
            ep_total += (m_score / CONSTANTS["STEPS_PER_MATCH"])

        scores.append(ep_total / 5) # Moyenne des 5 matchs

    return np.mean(scores[-50:]) # Moyenne des 50 derniers Ã©pisodes

# ==================== 5. MOTEUR GRID SEARCH ====================
def run_grid_search():
    keys, values = zip(*PARAM_GRID.items())
    combinations = [dict(zip(keys, v)) for v in itertools.product(*values)]
    total = len(combinations)

    print(f"ðŸ“Š OPTIMISATION Ã‰VOLUTIVE : {total} profils Ã  tester.")
    print("-" * 60)

    best_score = -float('inf')
    best_params = None

    for i, params in enumerate(combinations):
        start = time.time()
        score = evaluate_config(params)
        duration = time.time() - start

        print(f"[{i+1}/{total}] Score: {score:.3f} | {duration:.0f}s | {params}")

        if score > best_score:
            best_score = score
            best_params = params
            print("   >>> NOUVEAU LEADER ! ðŸ†")

    print("\n" + "="*60)
    print(f"MEILLEUR SCORE : {best_score:.3f}")
    print("CONFIG Ã€ COPIER DANS LE CODE PRINCIPAL :")
    print(best_params)

if __name__ == "__main__":
    run_grid_search()
```

```python
"""
COMPARAISON DQN : BASELINE vs AVEC PROBA (V2.1)
- Code trÃ¨s proche de ton V2.0
- Baseline = state inclut opponent.id/4.0 (comme ton code)
- Proba = state inclut p_coop_estime (probabilitÃ© empirique de coopÃ©ration de lâ€™adversaire)
- On garde le mÃªme tournoi, mÃªmes bots, mÃªmes hyperparamÃ¨tres
- On trace 2 graphes (Score + Coop) pour chaque version afin de comparer

Copie-colle et exÃ©cute.
"""

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque, defaultdict
import matplotlib.pyplot as plt
import pandas as pd
import copy

# ==================== CONFIGURATION ====================
class Config:
    ACTION_COOP = 0
    ACTION_DEFECT = 1

    # DQN
    STATE_SIZE = 3
    HIDDEN_SIZE = 64
    LEARNING_RATE = 0.001
    GAMMA = 0.95
    MEMORY_SIZE = 50000
    BATCH_SIZE = 64

    EPSILON_START = 1.0
    EPSILON_END = 0.05
    EPSILON_DECAY = 0.9995

    TARGET_UPDATE = 100  # steps

    # SARSA
    SARSA_ALPHA = 0.1
    SARSA_GAMMA = 0.95
    SARSA_EPSILON = 0.1

    # Simulation
    NUM_EPISODES = 600
    STEPS_PER_MATCH = 50

config = Config()

PAYOFF = {
    (0, 0): (3, 3), (0, 1): (0, 5),
    (1, 0): (5, 0), (1, 1): (1, 1)
}

# ==================== BOTS ====================
class Competitor:
    def __init__(self, name, id_num):
        self.name = name
        self.id = id_num
    def reset(self): pass
    def act(self, opp_last): return random.choice([0, 1])

class RandomBot(Competitor):
    def act(self, opp_last): return random.choice([0, 1])

class AngelBot(Competitor):
    def act(self, opp_last): return 0

class DevilBot(Competitor):
    def act(self, opp_last): return 1

class TitForTatBot(Competitor):
    def act(self, opp_last):
        return 0 if opp_last is None else opp_last

class SarsaBot(Competitor):
    def __init__(self, name, id_num):
        super().__init__(name, id_num)
        self.q_table = defaultdict(lambda: [0.0, 0.0])
        self.last_state = (0, 0)
        self.last_action = 0

    def reset(self):
        self.last_state = (0, 0)
        self.last_action = 0

    def act(self, opp_last):
        current_state = (self.last_action, opp_last if opp_last is not None else 0)
        if random.random() < config.SARSA_EPSILON:
            action = random.choice([0, 1])
        else:
            qs = self.q_table[current_state]
            action = 0 if qs[0] > qs[1] else 1
        self.current_state_buffer = current_state
        self.current_action_buffer = action
        return action

    def learn(self, reward, next_opp_action):
        next_state = (self.current_action_buffer, next_opp_action)
        old_q = self.q_table[self.current_state_buffer][self.current_action_buffer]
        max_next_q = max(self.q_table[next_state])
        new_q = old_q + config.SARSA_ALPHA * (reward + config.SARSA_GAMMA * max_next_q - old_q)
        self.q_table[self.current_state_buffer][self.current_action_buffer] = new_q
        self.last_action = self.current_action_buffer
        self.last_state = next_state

# ==================== DQN (Target Network) ====================
class DQNAgent(nn.Module):
    def __init__(self):
        super(DQNAgent, self).__init__()

        self.policy_net = nn.Sequential(
            nn.Linear(config.STATE_SIZE, config.HIDDEN_SIZE),
            nn.ReLU(),
            nn.Linear(config.HIDDEN_SIZE, config.HIDDEN_SIZE),
            nn.ReLU(),
            nn.Linear(config.HIDDEN_SIZE, 2)
        )

        self.target_net = copy.deepcopy(self.policy_net)
        self.target_net.eval()

        self.optimizer = optim.Adam(self.policy_net.parameters(), lr=config.LEARNING_RATE)
        self.memory = deque(maxlen=config.MEMORY_SIZE)
        self.epsilon = config.EPSILON_START
        self.steps_done = 0

    def select_action(self, state):
        self.steps_done += 1
        if random.random() < self.epsilon:
            return random.choice([0, 1])
        with torch.no_grad():
            return self.policy_net(torch.FloatTensor(state)).argmax().item()

    def remember(self, s, a, r, ns):
        self.memory.append((s, a, r, ns))

    def train_step(self):
        if len(self.memory) < config.BATCH_SIZE:
            return

        batch = random.sample(self.memory, config.BATCH_SIZE)
        states, actions, rewards, next_states = zip(*batch)

        states_t = torch.FloatTensor(np.array(states))
        actions_t = torch.LongTensor(actions).unsqueeze(1)
        rewards_t = torch.FloatTensor(rewards)
        next_states_t = torch.FloatTensor(np.array(next_states))

        q_curr = self.policy_net(states_t).gather(1, actions_t).squeeze()

        with torch.no_grad():
            q_next = self.target_net(next_states_t).max(1)[0]

        expected_q = rewards_t + config.GAMMA * q_next

        loss = nn.SmoothL1Loss()(q_curr, expected_q)

        self.optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(self.policy_net.parameters(), 1.0)
        self.optimizer.step()

        self.epsilon = max(config.EPSILON_END, self.epsilon * config.EPSILON_DECAY)

        if self.steps_done % config.TARGET_UPDATE == 0:
            self.target_net.load_state_dict(self.policy_net.state_dict())

# ==================== TOURNOI (MODE BASELINE vs MODE PROBA) ====================
def run_tournament(mode="baseline", seed=0):
    """
    mode = "baseline"  -> state = [dqn_last, opp_last, opponent.id/4.0]
    mode = "proba"     -> state = [dqn_last, opp_last, p_coop_estime]
                          oÃ¹ p_coop_estime = (nb_coop_opp + 1) / (nb_total + 2)  (lissage)
    """
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)

    print(f"\nðŸš€ DÃ‰MARRAGE TOURNOI | mode = {mode}")

    dqn = DQNAgent()

    sarsa_bot = SarsaBot("SARSA", 4)
    static_bots = [
        AngelBot("Angel", 0),
        DevilBot("Devil", 1),
        TitForTatBot("TfT", 2),
        RandomBot("Random", 3)
    ]
    all_opponents = static_bots + [sarsa_bot]

    history_scores = {bot.name: [] for bot in all_opponents}
    history_scores["DQN"] = []

    history_coop = {bot.name: [] for bot in all_opponents}
    history_coop["DQN"] = []

    for episode in range(config.NUM_EPISODES):

        # --- PHASE 1 : entraÃ®nement SARSA contre bots statiques ---
        for sparring_partner in static_bots:
            sarsa_bot.reset(); sparring_partner.reset()
            s_last, p_last = 0, 0
            for _ in range(20):
                act_s = sarsa_bot.act(p_last)
                act_p = sparring_partner.act(s_last)
                r_s, _ = PAYOFF[(act_s, act_p)]
                sarsa_bot.learn(r_s, act_p)
                s_last, p_last = act_s, act_p

        # --- PHASE 2 : tournoi ---
        dqn_ep_score = 0
        dqn_ep_coop_count = 0
        total_steps_ep = 0

        for opponent in all_opponents:
            opponent.reset()

            opp_score = 0
            opp_coop_count = 0

            # Pour estimer p_coop adversaire (mode proba)
            opp_coop_seen = 0
            opp_total_seen = 0

            dqn_last, opp_last = 0, 0

            for step in range(config.STEPS_PER_MATCH):

                # ---- Construire l'Ã©tat selon le mode ----
                if mode == "baseline":
                    z = opponent.id / 4.0
                elif mode == "proba":
                    # Lissage Beta(1,1) => (coop+1)/(total+2) pour Ã©viter 0/1 trop vite
                    z = (opp_coop_seen + 1.0) / (opp_total_seen + 2.0)
                else:
                    raise ValueError("mode doit Ãªtre 'baseline' ou 'proba'")

                state = np.array([dqn_last, opp_last, z], dtype=np.float32)

                act_dqn = dqn.select_action(state)
                act_opp = opponent.act(dqn_last)

                r_dqn, r_opp = PAYOFF[(act_dqn, act_opp)]

                # Mettre Ã  jour la proba aprÃ¨s avoir vu l'action adverse
                opp_total_seen += 1
                if act_opp == 0:
                    opp_coop_seen += 1

                # next_state (mÃªme logique)
                if mode == "baseline":
                    z_next = opponent.id / 4.0
                else:
                    z_next = (opp_coop_seen + 1.0) / (opp_total_seen + 2.0)

                next_state = np.array([act_dqn, act_opp, z_next], dtype=np.float32)

                dqn.remember(state, act_dqn, r_dqn, next_state)
                dqn.train_step()

                dqn_last, opp_last = act_dqn, act_opp

                # MÃ©triques
                opp_score += r_opp
                dqn_ep_score += r_dqn

                if act_opp == 0:
                    opp_coop_count += 1
                if act_dqn == 0:
                    dqn_ep_coop_count += 1

                total_steps_ep += 1

            history_scores[opponent.name].append(opp_score / config.STEPS_PER_MATCH)
            history_coop[opponent.name].append(opp_coop_count / config.STEPS_PER_MATCH)

        history_scores["DQN"].append(dqn_ep_score / total_steps_ep)
        history_coop["DQN"].append(dqn_ep_coop_count / total_steps_ep)

        if (episode + 1) % 50 == 0:
            print(f"Ep {episode+1:3d} | DQN Score: {history_scores['DQN'][-1]:.2f} | DQN Coop: {history_coop['DQN'][-1]*100:.0f}%")

    return history_scores, history_coop

# ==================== PLOT COMPARAISON ====================
def plot_comparison(scores_base, coops_base, scores_proba, coops_proba):
    colors = {"Angel": "cyan", "Devil": "black", "TfT": "blue",
              "Random": "gray", "SARSA": "red", "DQN": "green"}

    fig, axs = plt.subplots(2, 2, figsize=(14, 10))

    # --- (0,0) Scores Baseline ---
    ax = axs[0, 0]
    for name, data in scores_base.items():
        smooth = pd.Series(data).rolling(window=20).mean()
        lw = 3 if name == "DQN" else 1.2
        ls = '-' if name == "DQN" else '--'
        ax.plot(smooth, label=name, color=colors.get(name), linewidth=lw, linestyle=ls)
    ax.set_title("Scores (Baseline: ID adversaire)")
    ax.set_ylabel("Score")
    ax.grid(True, alpha=0.3)
    ax.legend(loc="center right")

    # --- (0,1) Scores Proba ---
    ax = axs[0, 1]
    for name, data in scores_proba.items():
        smooth = pd.Series(data).rolling(window=20).mean()
        lw = 3 if name == "DQN" else 1.2
        ls = '-' if name == "DQN" else '--'
        ax.plot(smooth, label=name, color=colors.get(name), linewidth=lw, linestyle=ls)
    ax.set_title("Scores (Proba: p_coop estimÃ©e)")
    ax.set_ylabel("Score")
    ax.grid(True, alpha=0.3)
    ax.legend(loc="center right")

    # --- (1,0) Coop Baseline ---
    ax = axs[1, 0]
    for name, data in coops_base.items():
        smooth = pd.Series(data).rolling(window=20).mean()
        lw = 3 if name == "DQN" else 1.2
        ls = '-' if name == "DQN" else ':'
        ax.plot(smooth, label=name, color=colors.get(name), linewidth=lw, linestyle=ls)
    ax.set_title("CoopÃ©ration (Baseline: ID adversaire)")
    ax.set_ylabel("Taux (0-1)")
    ax.set_xlabel("Ã‰pisodes")
    ax.grid(True, alpha=0.3)
    ax.legend(loc="center right")

    # --- (1,1) Coop Proba ---
    ax = axs[1, 1]
    for name, data in coops_proba.items():
        smooth = pd.Series(data).rolling(window=20).mean()
        lw = 3 if name == "DQN" else 1.2
        ls = '-' if name == "DQN" else ':'
        ax.plot(smooth, label=name, color=colors.get(name), linewidth=lw, linestyle=ls)
    ax.set_title("CoopÃ©ration (Proba: p_coop estimÃ©e)")
    ax.set_ylabel("Taux (0-1)")
    ax.set_xlabel("Ã‰pisodes")
    ax.grid(True, alpha=0.3)
    ax.legend(loc="center right")

    plt.tight_layout()
    plt.show()

# ==================== MAIN ====================
if __name__ == "__main__":
    # Option : pour une comparaison propre, on garde le mÃªme seed
    SEED = 0

    # 1) Baseline (comme ton code)
    scores_base, coops_base = run_tournament(mode="baseline", seed=SEED)

    # 2) Avec Proba
    scores_proba, coops_proba = run_tournament(mode="proba", seed=SEED)

    # 3) Comparaison visuelle
    plot_comparison(scores_base, coops_base, scores_proba, coops_proba)
```
